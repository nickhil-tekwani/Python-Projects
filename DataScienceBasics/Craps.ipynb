Craps simulator that simulates a game of craps played n number of times, keeping track of how many games were won or loss on a certain amount of roles. Displays win percentage as well.
Currently set to play 10_000 games.

```python
import numpy as np

wins = {}
losses = {}

# plays n number of craps games
def play_craps(number_of_games):
    for i in range(number_of_games):
        tuple = one_game()
        result = tuple[1]
        count = tuple[0]
        if result == "win":
            if count in wins:
                wins[count] += 1
            elif count not in wins:
                wins[count] = 1
        elif result == "loss":
            if count in losses:
                losses[count] += 1
            elif count not in losses:
                losses[count] = 1
    print("wins and losses dictionary updated")
        
# returns a tuple with number of rolls and whether this game was a win or loss
def one_game():
    # do first roll
    tuple = roll_dice()
    dice_sum = tuple[0] + tuple[1]
    
    # covers first roll possibilities
    if dice_sum in {7,11}:
        return (1, "win")
    elif dice_sum in {2,3,12}:
        return (1, "loss")
    
    # start counter
    counter = 1
    
    while 1==1:
        counter = counter + 1
        # new roll
        newTuple = roll_dice()
        newDice_sum = newTuple[0] + newTuple[1]
        if newDice_sum == dice_sum:
            return (counter, "win")
        elif newDice_sum == 7:
            return (counter, "loss")
        
# rolls 2 dice
def roll_dice():
    rand1 = np.random.randint(1,7)
    rand2 = np.random.randint(1,7)
    return (rand1, rand2)    

play_craps(10_000)

# displays results of wins and losses dict's in user friendly way
def display_results():
    #title
    print("ROLLS \t WINS \t LOSSES")
    
    #1-20 section, if losses and/or wins dont exist for any number of rolls, it fills 0
    for i in range(1,21):
        if (i in wins) and (i in losses):
            print(i, "\t", wins[i], "\t", losses[i])
        elif (i not in wins) and (i in losses):
            print(i, "\t", 0, "\t", losses[i])
        elif (i in wins) and (i not in losses):
            print(i, "\t", wins[i], "\t", 0)
        elif (i not in wins) and (i not in losses):
            print(i, "\t", 0, "\t", 0)
            
    #20+ section, if losses and/or wins dont exist for rolls above 20, it fills 0 for this row
    win_keys = list(wins.keys())
    loss_keys = list(losses.keys())
    win_keys_filtered = [i for i in win_keys if i >= 20]
    loss_keys_filtered = [j for j in loss_keys if j >= 20]
    w = (len(win_keys_filtered) != 0)
    l = (len(loss_keys_filtered) != 0)
    
    if (w and l):
        print("20+", "\t", sum(win_keys_filtered), "\t", sum(loss_keys_filtered))
    if ((not w) and l):
        print("20+", "\t", 0, "\t", sum(loss_keys_filtered))
    if (w and (not l)):
        print("20+", "\t", sum(win_keys_filtered), "\t", 0)
    if ((not w) and (not l)):
        print("20+", "\t", sum(win_keys_filtered), "\t", sum(loss_keys_filtered))
    
    #probability of winning
    number_won = sum(list(wins.values()))
    total = number_won + sum(list(losses.values()))
    print("The probability of winning is ", number_won, "/", total, "=", (number_won/total))
    
display_results()
```
